```python
import requests
from zipfile import ZipFile
from pathlib import Path

URL = "https://drive.usercontent.google.com/download?id=1hVeZbGh-wbF_OJx8leWYFH29vmZ8zk6y&export=download&authuser=0&confirm=t&uuid=f01b6e94-eb31-4bc4-a560-c9194fe508a6&at=ALWLOp7MPKJDzwDbTVfwPc6RQPb6:1765517460978"
OUT_ZIP = "./VLM_Demo_2.mp4"

r = requests.get(URL, stream=True)
r.raise_for_status()

with open(OUT_ZIP, "wb") as f:
    for chunk in r.iter_content(1024*1024):
        f.write(chunk)

print(f"[i] Downloaded {OUT_ZIP}")
```

```python
import cv2
import os

video_path = "/root/VLM-high-resolution/VLM_Demo_2.mp4"
output_dir = "VLM_Demo_2_frames"
desired_fps = 1
os.makedirs(output_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)
original_fps = cap.get(cv2.CAP_PROP_FPS)
frame_interval = int(original_fps / desired_fps)

print("Original FPS:", original_fps)
print("Extracting at:", desired_fps, "FPS")
print("Frame interval:", frame_interval)

frame_id = 0
saved_id = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_id % frame_interval == 0:
        frame_path = os.path.join(output_dir, f"frame_{saved_id:05d}.jpg")
        cv2.imwrite(frame_path, frame)
        saved_id += 1

    frame_id += 1

cap.release()
print("Done extracting frames")
```
